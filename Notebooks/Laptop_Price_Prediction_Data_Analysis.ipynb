```python
df['x_res'] = df['x_res'].str.replace(',' , '').str.findall(r'(\d+\.?\d+)').apply(lambda x: x[0])

"""Code Breakdown:

1. The str.replace(',', '') function removes any commas from the values in the 'x_res' column.  
   - Example: "1,920" becomes "1920".

2. The .str.findall(r'(\d+\.?\d+)') function searches for numbers in the string. It captures both whole numbers and decimal numbers if present.  
   - Example: "1920x1080" becomes ['1920', '1080'].

3. The .apply(lambda x: x[0]) function takes the first number from the list and assigns it to the column.  
   - Example: ['1920', '1080'] â†’ '1920'.

"""
```

#### Memory

```python
df['Memory'].value_counts()
```

```python

# Convert column to string and clean up text
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')

# Split into first and second storage layers
new = df["Memory"].str.split("+", n=1, expand=True)

df["first"] = new[0].str.strip()
df["second"] = new[1]

# Identify storage types for first layer
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Identify storage types for second layer BEFORE converting to int
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if pd.notna(x) and "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if pd.notna(x) and "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if pd.notna(x) and "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if pd.notna(x) and "Flash Storage" in x else 0)

# **Extract only numeric values safely**
df["first"] = df["first"].str.extract(r'(\d+)')  # Extract digits only
df["second"] = df["second"].str.extract(r'(\d+)')

# Handle NaN values before conversion
df["first"] = df["first"].fillna(0).astype(int)
df["second"] = df["second"].fillna(0).astype(int)

# Compute final storage capacity
df["HDD"] = (df["first"] * df["Layer1HDD"]) + (df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"]) + (df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"]) + (df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"]) + (df["second"] * df["Layer2Flash_Storage"])

# Drop unnecessary columns
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
                 'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
                 'Layer2Flash_Storage'], inplace=True)
```
